```python
import requests 
from bs4 import BeautifulSoup 
	
def getdata(url): 
	r = requests.get(url) 
	return r.text 
	
htmldata = getdata("https://www.geeksforgeeks.org/") 
soup = BeautifulSoup(htmldata, 'html.parser') 
for item in soup.find_all('img'): 
	print(item['src'])
```

```python
import requests

URL = "https://www.billboard.com/wp-content/uploads/2023/06/BLACKPINK-2023-coachella-billboard-1548.jpg?w=942&h=623&crop=1"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(URL, headers=headers)

if response.status_code == 200:
    with open("image.jpg", "wb") as f:
        f.write(response.content)
    print("Image downloaded successfully.")
else:
    print(f"Failed to download image. Status code: {response.status_code}")
```

```python
import requests
from PIL import Image
from io import BytesIO

URL = "https://www.billboard.com/wp-content/uploads/2023/06/BLACKPINK-2023-coachella-billboard-1548.jpg?w=942&h=623&crop=1"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(URL, headers=headers)

if response.status_code == 200:
    # Open the image using Pillow
    img = Image.open(BytesIO(response.content))

    # Display the image
    img.show()
else:
    print(f"Failed to download image. Status code: {response.status_code}")
```

```python
import requests
from IPython.display import Image, display
from PIL import Image as PILImage
from io import BytesIO

URL = "https://kongbakpao.com/wp-content/uploads/2023/05/kbp_blackpink_banner2.png"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(URL, headers=headers)

if response.status_code == 200:
    # Open the image using Pillow
    img = PILImage.open(BytesIO(response.content))

    # Display the image in the notebook
    display(Image(data=response.content))
else:
    print(f"Failed to download image. Status code: {response.status_code}")
```
